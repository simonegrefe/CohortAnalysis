```python
%sql
--CREATE OR REPLACE TABLE workspace.bigquery_db_cohort_db.silver_cohort_analysis AS 
WITH 

first_purchase AS (
  SELECT customer_id,
        COUNT(*) AS total_orders,
        MIN(order_date) AS first_order_date 
FROM workspace.bigquery_db_cohort_db.ecom_orders
GROUP BY customer_id
),

second_purchase AS (
  SELECT eo.customer_id,
        MIN(eo.order_date) AS second_order_date
  FROM workspace.bigquery_db_cohort_db.ecom_orders eo
  JOIN first_purchase fp 
  ON eo.customer_id = fp.customer_id  
  WHERE eo.order_date > fp.first_order_date
  GROUP BY eo.customer_id

)


SELECT
    fp.customer_id,
    fp.total_orders,
    fp.first_order_date,
    sp.second_order_date,
    DATEDIFF(sp.second_order_date, fp.first_order_date) AS days_between_orders
FROM first_purchase fp
LEFT JOIN second_purchase sp
ON fp.customer_id = sp.customer_id
--ORDER BY fp.customer_id ASC 
ORDER BY fp.first_order_date DESC
;
```
