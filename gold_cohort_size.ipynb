```python
%sql
--CREATE OR REPLACE TABLE workspace.bigquery_db_cohort_db.gold_cohort_size AS
SELECT
    -- monthly cohort
    DATE_TRUNC('month', first_order_date)::DATE AS cohort_month, 
    --each customer's first purchase
    COUNT(first_order_date) AS new_customers

  FROM workspace.bigquery_db_cohort_db.silver_cohort_analysis
  GROUP BY cohort_month
  ORDER BY cohort_month
```
