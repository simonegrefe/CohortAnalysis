```python
%sql

--CREATE OR REPLACE TABLE workspace.bigquery_db_cohort_db.gold_repeat_purchase AS

SELECT 
-- monthly cohort of first purchase
    DATE_TRUNC('month', first_order_date)::DATE AS cohort_month, 
    COUNT(*) AS num_customer,

--count number of customers with 2+ orders  
    SUM(CASE WHEN total_orders >= 2 THEN 1 ELSE 0 END) AS 2nd_order,
    SUM(CASE WHEN total_orders >= 3 THEN 1 ELSE 0 END) AS 3rd_order,
    SUM(CASE WHEN total_orders >= 4 THEN 1 ELSE 0 END) AS 4th_order,

--percentage of repeat purchase rate
    ROUND(100.0 * SUM(CASE WHEN total_orders >= 2 THEN 1 ELSE 0 END) / COUNT(*),2) AS repeat_rate_2nd_order,
    ROUND(100.0 * SUM(CASE WHEN total_orders >= 3 THEN 1 ELSE 0 END) / COUNT(*),2) AS repeat_rate_3rd_order,
    ROUND(100.0 * SUM(CASE WHEN total_orders >= 4 THEN 1 ELSE 0 END) / COUNT(*),2) AS repeat_rate_4th_order


FROM workspace.bigquery_db_cohort_db.silver_cohort_analysis
GROUP BY cohort_month
Order By cohort_month 
```
